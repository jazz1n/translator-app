```python
from transformers import BertTokenizer, DistilBertTokenizer, AlbertTokenizer
```

```python
text = "Tokenization is crucial for understanding how models process language."
```

```python
#different models
bert_tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
distilbert_tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased")
albert_tokenizer = AlbertTokenizer.from_pretrained("albert-base-v2")
```

```python
# Tokenizing
bert_tokens = bert_tokenizer.tokenize(text)
distilbert_tokens = distilbert_tokenizer.tokenize(text)
albert_tokens = albert_tokenizer.tokenize(text)
```

```python
print("BERT Tokenized Output:")
print(bert_tokens)
print(f"Number of tokens: {len(bert_tokens)}\n")
```

```python
print("DistilBERT Tokenized Output:")
print(distilbert_tokens)
print(f"Number of tokens: {len(distilbert_tokens)}\n")
```

```python
print("ALBERT Tokenized Output:")
print(albert_tokens)
print(f"Number of tokens: {len(albert_tokens)}\n")
```
